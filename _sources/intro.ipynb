# Solar Radiation Stations Catalog




```python
import pandas as pd
df = pd.read_csv('radiationstations.csv', sep=';', encoding='latin1', index_col='Abbreviation')
df
```

```python
import folium

# Initialize Folium map
m = folium.Map(
    location=[0, 15],
    zoom_start=1,
    min_zoom=1,
    max_bounds=True,
    tiles='openstreetmap',
    )

# Add each station to the map
for index, row in df.iterrows():
    folium.Marker(
        location=[row['Latitude'], row['Longitude']],
        popup=f"{row['Station full name']} ({row.name})",
        icon=folium.Icon(color='blue')
    ).add_to(m)

m  # Show the map
```
